```python
import random
from collections import defaultdict
from datasets import load_dataset, Dataset, DatasetDict
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer, DataCollatorWithPadding, EarlyStoppingCallback
import evaluate
import re
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sample_data(dataset, n_per_class, label_column="label", seed=42):
    random.seed(seed)
    label_buckets = defaultdict(list)

    for example in dataset:
        label_buckets[example[label_column]].append(example)

    sampled = []
    for label, examples in label_buckets.items():
        # Shuffle examples for the current label before extending
        random.shuffle(examples)
        sampled.extend(random.sample(examples, min(n_per_class, len(examples))))

    # Crucially, shuffle the final 'sampled' list before returning if you don't shuffle later
    random.shuffle(sampled) # Added this for robust mixing of classes
    return sampled
```
